```python
import sys

def calexpress(posix):
    tmp = 1
    stack = []
    answer = 0

    for i in range(len(posix)):
        if posix[i] == '(':
            stack.append(posix[i])
            tmp *= 2
        elif posix[i] == '[':
            stack.append(posix[i])
            tmp *= 3

        elif posix[i] == ")":
            if not stack or stack[-1] == "[":
                return 0
            if posix[i-1] == "(":
                answer += tmp
            stack.pop()
            tmp //= 2
        else:
            if not stack or stack[-1] == "(":
                return 0
            if posix[i-1] == "[":
                answer += tmp
            stack.pop()
            tmp //= 3
    return answer


sys.stdin = open('./input.txt')
posix = sys.stdin.readline().strip()
def validcheck(posix):
    stack = []
    # Check Bracket
    for ch in posix:
        if ch == '(':
            stack.append('(')
        elif ch == '[':
            stack.append('[')
        elif ch == ')':
            if stack and stack[-1] == '(':
                stack.pop(-1)
            else:
                return 0
        else:
            if stack and stack[-1] == '[':
                stack.pop(-1)
            else:
                return 0

    if stack:
        return 0

if validcheck(posix) != 0:
    result = calexpress(posix)
    print(result)
else:
    print(0)
```
