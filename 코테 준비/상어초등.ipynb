이제 학생의 만족도를 구해야 한다. 학생의 만족도는 자리 배치가 모두 끝난 후에 구할 수 있다.   
학생의 만족도를 구하려면 그 학생과 인접한 칸에 앉은 좋아하는 학생의 수를 구해야 한다. 그 값이 0이면 학생의 만족도는 0, 1이면 1, 2이면 10, 3이면 100, 4이면 1000이다.

학생의 만족도의 총 합을 구해보자.

```python
import sys
from collections import deque
sys.stdin = open("./input.txt", "r")

def satisfaction(sat, seat, dx, dy):
    #0, 1이면 1, 2이면 10, 3이면 100, 4이면 1000
    answer = 0
    score = [1,1, 10,100,1000]
    for i in range(len(sat)):
        for j in range(len(sat[i])):
            count = 0
            for k in range(4):
                nx = i + dx[k]
                ny = j + dy[k]
                if nx >= 0 and ny >=0 and nx < n and ny < n:
                    if sat[nx][ny] in seat[sat[i][j]]:
                        count += 1
            answer += score[count]
    return answer    

n = int(sys.stdin.readline())
seat = {} #좋아하는 사람 dict
sit_num = [] #앉는 순서
array = []
sat = [[False for _ in range(n)] for _ in range(n)] #앉은지 여부 체크

for i in range(n*n):
    temp = list(map(int, sys.stdin.readline().split()))
    seat[temp[0]] = temp[1:]
    sit_num.append(temp[0])
    array.append(temp)


sat[n//2][n//2] = sit_num[0]

dx = [0,0,-1,1]
dy = [1,-1,0,0]

for i in range(1, len(sit_num)):
    visited = [[False for _ in range(n)] for _ in range(n)]
    dq = deque([(n//2,n//2)])
    while dq:
        x,y = dq.popleft()
        if sat[x][y] in seat[sit_num[i]]: #x,y 좌표가 좋아하는 사람이 맞다면
            visited[x][y] = True
            point = []
            for j in range(4):
                nx = x + dx[j]
                ny = y + dy[j]
                if nx >= 0 and ny >=0 and nx < n and ny < n and sat[nx][ny] == False:
                    check = 0
                    like = 0
                    for l in range(4):
                        kx = nx + dx[l]
                        ky = ny + dy[l]
                        if kx >= 0 and ky >=0 and kx < n and ky < n:
                            if sat[kx][ky] == False:
                                check += 1
                            elif sat[kx][ky] in seat[sit_num[i]]:
                                like += 1
                    if len(point) == 0:
                        point.append([nx, ny, check, like])
                    else:
                        if point[0][3] < like:
                            point.pop()
                            point.append([nx, ny, check,like]) 
                        elif point[0][3] == like:
                            if point[0][2] < check:
                                point.pop()
                                point.append([nx, ny, check,like]) 
                            elif point[0][2] == check:
                                if point[0][0] > nx:
                                    point.pop()
                                    point.append([nx,ny, check, like]) 
                                elif point[0][0] == nx:
                                    if point[0][1] > ny:
                                        point.pop()
                                        point.append([nx,ny, check,like]) 
            
            if len(point) == 0:
                for j in range(4):
                    nx = x + dx[j]
                    ny = y + dy[j]
                    if nx >= 0 and ny >=0 and nx < n and ny < n and visited[nx][ny] == False:
                        dq.append((nx,ny))
            else:
                sat[point[0][0]][point[0][1]] = sit_num[i]
                break
                        
        else:
            for j in range(4):
                nx = x + dx[j]
                ny = y + dy[j]
                if nx >= 0 and ny >=0 and nx < n and ny < n and visited[nx][ny] == False:
                    dq.append((nx,ny))
                    
answer = satisfaction(sat, seat, dx,dy)
print(answer)
```
