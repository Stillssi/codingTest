```python
import sys
from collections import deque
sys.stdin = open('./input.txt', 'r')

n,m,k = map(int, sys.stdin.readline().split())
array = []
dx = [0,1,0,-1]
dy = [1,0,-1,0] #동, 남, 서, 북
dice = [1,2,3,4,5,6]
for i in range(n):
    temp = list(map(int, sys.stdin.readline().split()))
    array.append(temp)

def rollingDice(dice, direction, now_x, now_y): #현재 밑바닥 값, 방향(0:동, 1:남, 2 :서, 3:북), 현재 위치
    nx = now_x + dx[direction]
    ny = now_y + dy[direction]
    if nx < 0 or ny < 0 or nx >= n or ny >= m:
        if direction == 0:
            direction = 2
        elif direction == 1:
            direction = 3
        elif direction == 2:
            direction =0
        else:
            direction = 1
        nx = now_x + dx[direction]
        ny = now_y + dy[direction]
        
    a, b, c, d, e, f = dice[0], dice[1], dice[2], dice[3], dice[4], dice[5]
    if direction == 0: #동
        dice[0], dice[1], dice[2], dice[3], dice[4], dice[5] = d, b, a, f, e, c
    elif direction == 2: #서
        dice[0], dice[1], dice[2], dice[3], dice[4], dice[5] = c, b, f, a, e, d
    elif direction == 3: #북
        dice[0], dice[1], dice[2], dice[3], dice[4], dice[5] = e, a, c, d, f, b
    else:
        dice[0], dice[1], dice[2], dice[3], dice[4], dice[5] = b, f, c, d, a, e
    return dice[5], nx, ny, direction

def around_blank(now_x, now_y, array , num):
    visited = [[False for _ in range(len(array[0]))] for _ in range(len(array))]
    dq = deque([(now_x, now_y)])
    count = 0
    while dq:
        x,y=dq.popleft()
        for i in range(4):
            nx = x + dx[i]
            ny = y + dy[i]
            if nx >= 0 and ny >= 0 and nx < n and ny <m and visited[nx][ny] == False and array[nx][ny] == num:
                visited[nx][ny] = True
                count += 1
                dq.append((nx,ny))
    return count if count >= 1 else 1
    
x, y, direct = 0,0,0
answer = 0
for i in range(0, k):
    num, x, y, direct= rollingDice(dice,direct,x,y) #밑바닥, 위치x, 위치y
    blank = around_blank(x,y,array, array[x][y])
    if num > array[x][y]:
        direct += 1
        if direct > 3:
            direct = 0
    elif num < array[x][y]:
        direct -= 1
        if direct < 0:
            direct = 3        
    answer += blank * array[x][y]

print(answer)
```
